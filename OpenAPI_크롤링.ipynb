```
datas = []

for i in range(len(te)):
  lev = te[i].lev.string.strip()
  itemcode = te[i].msdsItemCode.string.strip()
  upitemcode = te[i].upMsdsItemCode.string.strip()
  itemname = te[i].msdsItemNameKor.string.strip()
  itemno = te[i].msdsItemNo
  order = te[i].ordrIdx.string.strip()
  detail = te[i].itemDetail

  data = [lev, itemcode, upitemcode, itemname, itemno, order, detail]
  datas.append(data)
```
