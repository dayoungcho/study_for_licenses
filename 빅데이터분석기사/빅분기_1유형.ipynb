### 6. 댓글의 수(comment_count)로 영상 반응에 대한 판단을 할 수 있다. viewcount 대비 댓글 수가 가장 높은 영상을 확인하라. ( view_count 값이 0인 경우는 제외한다.)

### 8. like 대비 dislike의 수가 가장 적은 영상은 무엇인가? (like, dislike 값이 0인 경우는 제외한다)

### 4. 각 채널의 2021-10-03 03:00:00 ~ 2021-11-01 15:00:00까지 구독자수(subcnt)의 증가량을 구하여라

```python
before = channel[channel.ct >= pd.to_datetime('2021-10-03 03:00:00')].sort_values(['ct','channelname']).drop_duplicates('channelname')[['channelname','subcnt']]
before
```

```python
after = channel[channel.ct <= pd.to_datetime('2021-11-01 15:00:00')].sort_values(['ct','channelname']).drop_duplicates('channelname', keep='last')[['channelname','subcnt']]
after
```

```python
num_4 = pd.merge(before, after, on='channelname')
num_4['del'] = num_4['subcnt_y'] - num_4['subcnt_x']
num_4 = num_4.drop(['subcnt_x','subcnt_y'],axis=1)
num_4
```

### 5. 각 비디오는 10분 간격으로 구독자수, 좋아요, 싫어요수, 댓글수가 수집된 것으로 알려졌다. 공범 EP1의 비디오정보 데이터 중 수집간격이 5분 이하, 20분 이상인 데이터 구간(해당 시점 전, 후)의 시각을 모두 출력하라. V 안할래..

```python
import datetime as dt

num_5 = video[video.videoname == ' 공범 EP1'].sort_values('ct').reset_index()
lst = list()
for i in range(len(num_5)):
    time = video.loc[i+1,'ct'] - video.loc[i,'ct']
    if (time <= dt.timedelta(minutes=5)) or (time >= dt.timedelta(minutes=20)):
        lst.append(i)
        lst.append(i+1)

video.loc[lst,:]
```

### 6. 각 에피소드의 시작날짜(년-월-일)를 에피소드 이름과 묶어 데이터프레임으로 만들고 출력하라

```python
num_6 = video.sort_values(['videoname', 'ct']).drop_duplicates('videoname')[['ct','videoname']]
num_6['date'] = num_6.ct.dt.date
num_6[['date','videoname']]
```

### 7. '공범' 컨텐츠의 경우 19:00시에 공개 되는 것으로 알려져 있다. 공개된 날의 21시의 viewcnt, ct, videoname으로 구성된 데이터프레임을 내림차순으로 정렬하여 출력하라.

```python
num_7 = video.sort_values(['videoname','ct'])
num_7 = num_7[num_7.ct.dt.hour==21].drop_duplicates('videoname').sort_values('viewcnt',ascending=False).reset_index(drop=True)
num_7[['videoname','viewcnt','ct']]
```

### 8. video 정보의 가장 최근 데이터들에서 각 에피소드의 싫어요/좋아요 비율을 ratio 컬럼으로 만들고 videoname, ratio로 구성된 데이터프레임을 ratio의 오름차순으로 정렬하라.

```python
num_8 = video.sort_values(['ct','videoname']).drop_duplicates('videoname',keep='last')
num_8['ratio'] = num_8.dislikecnt / num_8.likecnt
num_8 = num_8.sort_values('ratio').reset_index(drop=True)
num_8[['videoname','ratio']]
```

### 9. 2021-11-01 00:00:00 ~ 15:00:00까지 각 에피소드별 viewcnt의 증가량을 데이터프레임으로 만드시오.

```python
start = pd.to_datetime('2021-11-01 00:00:00')
end = pd.to_datetime('2021-11-01 15:00:00')
num_9 = video[(video.ct >= start) & (video.ct <= end)].sort_values(['videoname','ct'])

start_df = num_9.drop_duplicates('videoname').reset_index(drop=True)[['videoname','viewcnt']]
end_df = num_9.drop_duplicates('videoname',keep='last').reset_index(drop=True)[['videoname','viewcnt']]

num_9 = start_df.merge(end_df,on='videoname')
num_9['viewcnt'] = num_9.viewcnt_y - num_9.viewcnt_x
num_9[['videoname','viewcnt']]
```
