### 5. ê° ë¹„ë””ì˜¤ëŠ” 10ë¶„ ê°„ê²©ìœ¼ë¡œ êµ¬ë…ììˆ˜, ì¢‹ì•„ìš”, ì‹«ì–´ìš”ìˆ˜, ëŒ“ê¸€ìˆ˜ê°€ ìˆ˜ì§‘ëœ ê²ƒìœ¼ë¡œ ì•Œë ¤ì¡Œë‹¤. ê³µë²” EP1ì˜ ë¹„ë””ì˜¤ì •ë³´ ë°ì´í„° ì¤‘ ìˆ˜ì§‘ê°„ê²©ì´ 5ë¶„ ì´í•˜, 20ë¶„ ì´ìƒì¸ ë°ì´í„° êµ¬ê°„(í•´ë‹¹ ì‹œì  ì „, í›„)ì˜ ì‹œê°ì„ ëª¨ë‘ ì¶œë ¥í•˜ë¼. V ì•ˆí• ë˜..

```python
import datetime as dt

num_5 = video[video.videoname == ' ê³µë²” EP1'].sort_values('ct').reset_index()
lst = list()
for i in range(len(num_5)):
    time = video.loc[i+1,'ct'] - video.loc[i,'ct']
    if (time <= dt.timedelta(minutes=5)) or (time >= dt.timedelta(minutes=20)):
        lst.append(i)
        lst.append(i+1)

video.loc[lst,:]
```

### 9. 2021-11-01 00:00:00 ~ 15:00:00ê¹Œì§€ ê° ì—í”¼ì†Œë“œë³„ viewcntì˜ ì¦ê°€ëŸ‰ì„ ë°ì´í„°í”„ë ˆì„ìœ¼ë¡œ ë§Œë“œì‹œì˜¤.

```python
start = pd.to_datetime('2021-11-01 00:00:00')
end = pd.to_datetime('2021-11-01 15:00:00')
num_9 = video[(video.ct >= start) & (video.ct <= end)].sort_values(['videoname','ct'])

start_df = num_9.drop_duplicates('videoname').reset_index(drop=True)[['videoname','viewcnt']]
end_df = num_9.drop_duplicates('videoname',keep='last').reset_index(drop=True)[['videoname','viewcnt']]

num_9 = start_df.merge(end_df,on='videoname')
num_9['viewcnt'] = num_9.viewcnt_y - num_9.viewcnt_x
num_9[['videoname','viewcnt']]
```
