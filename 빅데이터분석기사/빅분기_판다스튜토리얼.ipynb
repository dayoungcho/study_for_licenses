### 57. Income_Category의 카테고리를 map 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라   V
Unknown : N
Less than $40K : a
$40K - $60K : b
$60K - $80K : c
$80K - $120K : d
$120K +’ : e

```python
dic = {'Unknown':'N', 
       'Less than $40K' : 'a', 
       '$40K - $60K' : 'b', 
       '$60K - $80K' : 'c', 
       '$80K - $120K' : 'd', 
       '$120K +' : 'e'}

num_57 = df.copy()
num_57['newIncome'] = num_57.Income_Category.map(lambda x: dic[x])
num_57
```

### 58. Income_Category의 카테고리를 apply 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라 Unknown : N
Less than $40K : a
$40K - $60K : b
$60K - $80K : c
$80K - $120K : d
$120K +’ : e

```python
num_58 = df.copy()
dic = {'Unknown':'N', 
       'Less than $40K' : 'a', 
       '$40K - $60K' : 'b', 
       '$60K - $80K' : 'c', 
       '$80K - $120K' : 'd', 
       '$120K +' : 'e'}

num_58['newIncome'] = num_58.Income_Category.apply(lambda x: dic[x])
num_58
```

```python
# 풀이

num_58 = df.copy()
def change_cat(x):
    if x =='Unknown':
        return 'N'
    elif x =='Less than $40K':
        return 'a'
    elif x =='$40K - $60K':   
        return 'b'
    elif x =='$60K - $80K':    
        return 'c'
    elif x =='$80K - $120K':   
        return 'd'
    elif x =='$120K +' :     
        return 'e'

num_58['newIncome'] = num_58.Income_Category.apply(change_cat)

num_58
```

### 59. Customer_Age의 값을 이용하여 나이 구간을 AgeState 컬럼으로 정의하라. (0~9 : 0 , 10~19 :10 , 20~29 :20 … 각 구간의 빈도수를 출력하라

```python
df.Customer_Age.max()
```

```python
num_59 = df.copy()

def age_cat(x):
    if x >= 20 and x < 30:
        return 20
    elif x >= 30 and x < 40:
        return 30
    elif x >= 40 and x < 50:
        return 40
    elif x >= 50 and x < 60:
        return 50
    elif x >= 60 and x < 70:
        return 60
    else:
        return 70

num_59['AgeState'] = num_59.Customer_Age.apply(age_cat)
num_cat = num_59.AgeState.value_counts()
num_cat.sort_index()
```

```python
# 풀이

num_59 = df.copy()
num_59['AgeState'] = num_59.Customer_Age.map(lambda x: x//10 * 10)
Ans = num_59['AgeState'].value_counts().sort_index()
Ans
```

### 62. Marital_Status 컬럼값이 Married 이고 Card_Category 컬럼의 값이 Platinum인 경우 1 그외의 경우에는 모두 0으로 하는 newState컬럼을 정의하라. newState의 각 값들의 빈도수를 출력하라 V

```python
num_62 = df.copy()

def check(x):
    if x.Marital_Status == 'Married' and x.Card_Category == 'Platinum':
        return 1
    else:
        return 0
    
num_62['newState'] = num_62.apply(check, axis=1)

num_62.newState.value_counts()
```
